```python
import cv2
```

```python
face_cascade=cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_frontalface_default.xml")
smile_cascade=cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_smile.xml")
```

```python
def detect_smile(gray,frame):
    faces=face_cascade.detectMultiScale(gray,scaleFactor=1.3,minNeighbors=5,minSize=(30,30))
    
    for x,y,w,h in faces:
        roi_gray=gray[y:y+h,x:x+w]
        roi_color=frame[y:y+h,x:x+w]
        
        smiles=smile_cascade.detectMultiScale(roi_color,scaleFactor=1.8,minNeighbors=20,minSize=(25,25))
        for sx,sy,sw,sh in smiles:
            cv2.rectangle(roi_color,(sx,sy),(sx+sw,sy+sh),(255,0,0),2)
    return frame
```

```python
frame=cv2.imread("smile.jpg")
gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
canvas=detect_smile(gray,frame)
cv2.imshow("smile detection",canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
X_train, X_test = X_train / 255.0, X_test / 255.0
```

```python
X_train_flat, X_test_flat = X_train.reshape((-1, 28*28)), X_test.reshape((-1, 28*28))
```

```python
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(64, activation="relu"),
    keras.layers.Dense(32, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
])
model.compile(loss="sparse_categorical_crossentropy", optimizer="sgd", metrics=["accuracy"])
```

```python
loss = model.fit(X_train, y_train, epochs=20, validation_split=0.2)
```

```python
pd.DataFrame(loss.history).plot()
```

```python
accuracy = model.evaluate(X_test, y_test)[1]
print(f'Accuracy on test set: {accuracy:.2%}')
```

```python
plt.figure(figsize=(10, 10))
for label in range(10):
    indices = np.where(model.predict(X_test).argmax(axis=1) == label)[0]
    for i in range(10):
        plt.subplot(10, 10, label * 10 + i + 1)
        plt.imshow(X_test[indices[i]], cmap='gray')
        plt.axis('off')
plt.show()
```
